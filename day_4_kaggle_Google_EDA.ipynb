```python
import pandas as pd

# Sample data
data = {'Size': [
    'Varies with device', '11M', '12M', '13M', '14M',
    '429k', '200k', '460k', '728k', '619k'
]}

# Create a DataFrame from the sample data
df = pd.DataFrame(data)

# Define a function to convert the values to numeric
def convert_size(size):
    # Check if the value is a string (not a float)
    if isinstance(size, str):
        # If the value contains 'M', remove 'M' and convert to float
        if 'M' in size:
            return float(size.replace('M', ''))
        # If the value contains 'k', remove 'k', convert to float, and divide by 1000
        elif 'k' in size:
            return float(size.replace('k', '')) / 1000
        # If the value is 'Varies with device', return None
        elif size == 'Varies with device':
            return None
    # For non-string values (e.g., float), return None
    return None

# Apply the conversion function to the 'Size' column
df['Size'] = df['Size'].apply(convert_size)

# Now, the 'Size' column contains numeric values
print(df)
```

```python
df['Size'].value_counts()
```

```python
import pandas as pd

# Sample data
data = {
    'Installs': [
        '1,000,000+',
        '10,000,000+',
        '100,000+',
        '10,000+',
        '1,000+',
        '5,000,000+',
        '100+',
        '500,000+',
        '50,000+',
        '5,000+',
        '100,000,000+',
        '10+',
        '500+',
        '50,000,000+',
        '50+',
        '5+',
        '500,000,000+',
        '1+',
        '1,000,000,000+',
        '0+',
        '0',
        'Free'
    ]
}

df = pd.DataFrame(data)

# Define a function to convert non-numeric values to numeric
def convert_to_numeric(value):
    try:
        # Extract the numeric part from the string and convert to integer
        return int(''.join(filter(str.isdigit, value)))
    except:
        return 0  # Treat non-numeric values as 0

# Apply the conversion function to the 'Installs' column
df['Installs'] = df['Installs'].apply(convert_to_numeric)

# Now, the 'Installs' column contains numeric values
print(df)
```

```python
df['Installs'].value_counts()
```

```python
import pandas as pd

# Sample data
data = {
    'Price': [
        '0',
        '$0.99',
        '$2.99',
        '$1.99',
        '$4.99',
        '$1.75',
        '$14.00',
        '$4.85',
        '$46.99',
        '$1.04',
    ]
}

df = pd.DataFrame(data)

# Define a function to convert non-numeric values to numeric
def convert_price_to_numeric(value):
    try:
        # Remove the dollar sign and convert to float
        return float(value.lstrip('$'))
    except:
        return 0.0  # Treat other non-numeric values as 0

# Apply the conversion function to the 'Price' column
df['Price'] = df['Price'].apply(convert_price_to_numeric)

# Now, the 'Price' column contains numeric values
print(df)
```

```python
df['Price'].value_counts()
```
